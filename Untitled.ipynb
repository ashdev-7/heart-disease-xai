```python
# Cell 3: Preprocessing — Imputation, Scaling, SMOTE (train only)

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

RANDOM_SEED = 42

# ── Feature type definitions ──────────────────────────────────────────────────
TARGET       = "TenYearCHD"

# Mode imputation
CATEGORICAL  = ["male", "education", "currentSmoker",
                "BPMeds", "prevalentStroke", "prevalentHyp", "diabetes"]

# Median imputation (skewed)
MEDIAN_COLS  = ["glucose"]

# Mean imputation (reasonably symmetric)
MEAN_COLS    = ["cigsPerDay", "totChol", "BMI", "heartRate"]

# BPMeds is binary float — treat as categorical for imputation
# (already in CATEGORICAL above)

print("=" * 55)
print("PREPROCESSING PIPELINE")
print("=" * 55)

# ── 1. Separate features and target ──────────────────────────────────────────
X = df_raw.drop(columns=[TARGET])
y = df_raw[TARGET]

print(f"\nFeatures : {X.shape[1]}  |  Samples : {X.shape[0]}")

# ── 2. Stratified 80/20 split — before any imputation ────────────────────────
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=RANDOM_SEED
)

print(f"\nTrain raw : {X_train_raw.shape[0]} rows")
print(f"Test  raw : {X_test_raw.shape[0]} rows")
print(f"\nTrain class distribution:")
print(f"  Class 0 : {(y_train==0).sum()}  ({(y_train==0).mean()*100:.1f}%)")
print(f"  Class 1 : {(y_train==1).sum()}  ({(y_train==1).mean()*100:.1f}%)")

# ── 3. Imputation — fit on train, apply to both ───────────────────────────────
X_train = X_train_raw.copy()
X_test  = X_test_raw.copy()

imputation_log = {}

for col in CATEGORICAL:
    if X_train[col].isnull().any() or X_test[col].isnull().any():
        val = X_train[col].mode()[0]
        X_train[col].fillna(val, inplace=True)
        X_test[col].fillna(val,  inplace=True)
        imputation_log[col] = ("mode", val)

for col in MEDIAN_COLS:
    if X_train[col].isnull().any() or X_test[col].isnull().any():
        val = X_train[col].median()
        X_train[col].fillna(val, inplace=True)
        X_test[col].fillna(val,  inplace=True)
        imputation_log[col] = ("median", val)

for col in MEAN_COLS:
    if X_train[col].isnull().any() or X_test[col].isnull().any():
        val = X_train[col].mean()
        X_train[col].fillna(val, inplace=True)
        X_test[col].fillna(val,  inplace=True)
        imputation_log[col] = ("mean", round(val, 4))

print("\nImputation applied (fitted on train only):")
for col, (method, val) in imputation_log.items():
    print(f"  {col:<20} {method:<8}  fill value = {val}")

# Verify no nulls remain
assert X_train.isnull().sum().sum() == 0, "Nulls remain in train"
assert X_test.isnull().sum().sum()  == 0, "Nulls remain in test"
print("\nNull check passed — no missing values in train or test.")

# ── 4. StandardScaler — fit on train only ─────────────────────────────────────
scaler      = StandardScaler()
X_train_sc  = pd.DataFrame(
    scaler.fit_transform(X_train),
    columns=X_train.columns, index=X_train.index
)
X_test_sc   = pd.DataFrame(
    scaler.transform(X_test),
    columns=X_test.columns,  index=X_test.index
)
print("\nStandardScaler fitted on train, applied to test.")

# ── 5. SMOTE — applied to scaled train only ───────────────────────────────────
# Test set is NEVER touched. SMOTE operates in scaled space for consistency
# with what the models will see during training.
smote = SMOTE(random_state=RANDOM_SEED, k_neighbors=5)
X_train_bal_arr, y_train_bal_arr = smote.fit_resample(X_train_sc, y_train)

X_train_bal = pd.DataFrame(X_train_bal_arr, columns=X_train_sc.columns)
y_train_bal = pd.Series(y_train_bal_arr, name=TARGET)

print("\nSMOTE applied to training set only:")
print(f"  Before — Class 0: {(y_train==0).sum()}  "
      f"Class 1: {(y_train==1).sum()}  "
      f"Total: {len(y_train)}")
print(f"  After  — Class 0: {(y_train_bal==0).sum()}  "
      f"Class 1: {(y_train_bal==1).sum()}  "
      f"Total: {len(y_train_bal)}")
print(f"  Synthetic samples added: "
      f"{len(y_train_bal) - len(y_train)}")

print(f"\nFinal shapes:")
print(f"  X_train_bal : {X_train_bal.shape}  (SMOTE-balanced, for model training)")
print(f"  X_test_sc   : {X_test_sc.shape}    (original patients, for evaluation)")
print(f"  y_train_bal : {y_train_bal.shape}")
print(f"  y_test      : {y_test.shape}")

print("\nPreprocessing complete.")
```

```python
# Cell 5: SHAP Explainability
# LinearExplainer → LR (exact)
# TreeExplainer   → RF (exact)
# KernelExplainer → MLP (approximation, 100 real background samples)

import shap
import numpy as np
import pandas as pd
import torch
import warnings
warnings.filterwarnings("ignore")

RANDOM_SEED = 42
N_BACKGROUND = 100   # KernelExplainer background size
N_STABILITY  = 30    # instances for stability evaluation

print("=" * 60)
print("CELL 5 — SHAP EXPLAINABILITY")
print("=" * 60)

# ── 30 fixed stability instances from real test set ───────────────────────────
stability_idx      = X_test_sc.sample(n=N_STABILITY,
                                       random_state=RANDOM_SEED).index
X_stability        = X_test_sc.loc[stability_idx]
y_stability        = y_test.loc[stability_idx]

print(f"\n  Stability instances: {N_STABILITY} fixed test patients")
print(f"  Class dist in stability set: "
      f"{(y_stability==0).sum()} neg, {(y_stability==1).sum()} pos")

shap_store = {
    "global_importance"   : {},   # mean |SHAP| per feature — for alignment
    "stability_shap"      : {},   # per-instance SHAP for 30 instances
    "full_test_shap"      : {},   # all test instances — for global ranking
}

# ═══════════════════════════════════════════════════════
# 1. LR — LinearExplainer
# ═══════════════════════════════════════════════════════
print("\n── 1. LR — LinearExplainer (exact) ──")
lr_model    = final_results["LR"]["model"]
explainer_lr = shap.LinearExplainer(lr_model, X_train_sc)

shap_lr_full  = explainer_lr.shap_values(X_test_sc)     # (848, 15)
shap_lr_stab  = explainer_lr.shap_values(X_stability)   # (30, 15)

# Global importance: mean absolute SHAP across all test instances
lr_global = pd.Series(
    np.abs(shap_lr_full).mean(axis=0),
    index=X_test_sc.columns
).sort_values(ascending=False)

shap_store["global_importance"]["LR"]  = lr_global
shap_store["stability_shap"]["LR"]     = shap_lr_stab
shap_store["full_test_shap"]["LR"]     = shap_lr_full

print(f"  Done. Shape: {shap_lr_full.shape}")
print(f"  Top 5 features by mean |SHAP|:")
for feat, val in lr_global.head(5).items():
    print(f"    {feat:<20} {val:.5f}")

# ═══════════════════════════════════════════════════════
# 2. RF — TreeExplainer (exact Shapley values through tree)
# ═══════════════════════════════════════════════════════
print("\n── 2. RF — TreeExplainer (exact) ──")
rf_model     = final_results["RF"]["model"]
explainer_rf = shap.TreeExplainer(rf_model)

shap_rf_full_raw = explainer_rf.shap_values(X_test_sc)
shap_rf_stab_raw = explainer_rf.shap_values(X_stability)

# RF returns list [class0, class1] or 3D array — extract positive class
if isinstance(shap_rf_full_raw, list):
    shap_rf_full = shap_rf_full_raw[1]
    shap_rf_stab = shap_rf_stab_raw[1]
elif shap_rf_full_raw.ndim == 3:
    shap_rf_full = shap_rf_full_raw[:, :, 1]
    shap_rf_stab = shap_rf_stab_raw[:, :, 1]
else:
    shap_rf_full = shap_rf_full_raw
    shap_rf_stab = shap_rf_stab_raw

rf_global = pd.Series(
    np.abs(shap_rf_full).mean(axis=0),
    index=X_test_sc.columns
).sort_values(ascending=False)

shap_store["global_importance"]["RF"]  = rf_global
shap_store["stability_shap"]["RF"]     = shap_rf_stab
shap_store["full_test_shap"]["RF"]     = shap_rf_full

print(f"  Done. Shape: {shap_rf_full.shape}")
print(f"  Top 5 features by mean |SHAP|:")
for feat, val in rf_global.head(5).items():
    print(f"    {feat:<20} {val:.5f}")

# ═══════════════════════════════════════════════════════
# 3. MLP — KernelExplainer (approximation)
# Background: 100 samples from ORIGINAL (non-SMOTE) train set
# Rationale: background should represent real patient distribution
#            not the synthetic 50/50 SMOTE distribution
# ═══════════════════════════════════════════════════════
print("\n── 3. MLP — KernelExplainer (approximation) ──")
print(f"  Background: {N_BACKGROUND} samples from original "
      f"(non-SMOTE) train set")
print("  This will take several minutes on 848 test instances...")

mlp_model = final_results["MLP"]["model"]
mlp_model.eval()

def mlp_predict(x):
    """Wrapper: numpy array → probability scalar per instance."""
    x_t = torch.FloatTensor(x)
    with torch.no_grad():
        return torch.sigmoid(mlp_model(x_t)).numpy().flatten()

# Background from real (non-SMOTE) scaled training data
background = shap.sample(X_train_sc, N_BACKGROUND, random_state=RANDOM_SEED)
explainer_mlp = shap.KernelExplainer(mlp_predict, background)

# Full test set SHAP (slow — ~5-10 min on H100 for 848 instances)
print("  Computing full test set SHAP values...")
shap_mlp_full = explainer_mlp.shap_values(X_test_sc, silent=False)

# Stability instances SHAP
print("  Computing stability instance SHAP values (30 instances)...")
shap_mlp_stab = explainer_mlp.shap_values(X_stability, silent=True)

mlp_global = pd.Series(
    np.abs(shap_mlp_full).mean(axis=0),
    index=X_test_sc.columns
).sort_values(ascending=False)

shap_store["global_importance"]["MLP"] = mlp_global
shap_store["stability_shap"]["MLP"]    = shap_mlp_stab
shap_store["full_test_shap"]["MLP"]    = shap_mlp_full

print(f"\n  Done. Shape: {np.array(shap_mlp_full).shape}")
print(f"  Top 5 features by mean |SHAP|:")
for feat, val in mlp_global.head(5).items():
    print(f"    {feat:<20} {val:.5f}")

# ═══════════════════════════════════════════════════════
# 4. Cross-model top-5 comparison (quick sanity check)
# ═══════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("TOP-5 FEATURE RANKING COMPARISON")
print("=" * 60)
print(f"  {'Rank':<6} {'LR (SHAP)':<22} {'RF (SHAP)':<22} {'MLP (SHAP)':<22}")
print(f"  {'-'*5} {'-'*21} {'-'*21} {'-'*21}")
lr_top  = lr_global.index.tolist()
rf_top  = rf_global.index.tolist()
mlp_top = mlp_global.index.tolist()
for i in range(5):
    print(f"  {i+1:<6} {lr_top[i]:<22} {rf_top[i]:<22} {mlp_top[i]:<22}")

# ── Jaccard top-5 vs LR anchor ────────────────────────────────────────────────
def jaccard(s1, s2):
    s1, s2 = set(s1), set(s2)
    return len(s1 & s2) / len(s1 | s2)

j_rf_3  = jaccard(lr_top[:3], rf_top[:3])
j_mlp_3 = jaccard(lr_top[:3], mlp_top[:3])
j_rf_5  = jaccard(lr_top[:5], rf_top[:5])
j_mlp_5 = jaccard(lr_top[:5], mlp_top[:5])

print(f"\n  Jaccard similarity vs LR anchor:")
print(f"    {'':6} {'Top-3':>8}  {'Top-5':>8}")
print(f"    {'RF':6} {j_rf_3:>8.4f}  {j_rf_5:>8.4f}")
print(f"    {'MLP':6} {j_mlp_3:>8.4f}  {j_mlp_5:>8.4f}")

print("\n  SHAP generation complete. shap_store populated.")
print("  Proceed to Cell 6 (Integrated Gradients).")
```

```python
# Cell 6: Integrated Gradients — MLP only
# Baseline: training mean vector (real patient average)
# Steps: 50 (standard for tabular data)
# Completeness check: attributions should sum to
#   f(x) - f(baseline) for each instance

import torch
import numpy as np
import pandas as pd
from captum.attr import IntegratedGradients
import warnings
warnings.filterwarnings("ignore")

print("=" * 60)
print("CELL 6 — INTEGRATED GRADIENTS (MLP only)")
print("=" * 60)

# ── Setup ─────────────────────────────────────────────────────────────────────
mlp_model = final_results["MLP"]["model"]
mlp_model.eval()

class MLPWrapper(torch.nn.Module):
    """Wraps MLP to output probability (sigmoid) for IG."""
    def __init__(self, model):
        super().__init__()
        self.model = model
    def forward(self, x):
        return torch.sigmoid(self.model(x))

wrapper = MLPWrapper(mlp_model)
ig      = IntegratedGradients(wrapper)

# ── Baseline: training mean vector ────────────────────────────────────────────
# PRD §5.2: mean of ORIGINAL (non-SMOTE) training set
# Rationale: represents the "average real patient"
#            zero baseline is physiologically meaningless
baseline_np     = X_train_sc.mean(axis=0).values.astype(np.float32)
baseline_tensor = torch.FloatTensor(baseline_np).unsqueeze(0)

print(f"\n  Baseline: training mean vector (original, non-SMOTE)")
print(f"  Integration steps: 50")
print(f"  Model output at baseline: "
      f"{torch.sigmoid(mlp_model(baseline_tensor)).item():.4f}")

# ── Global IG — all test instances ────────────────────────────────────────────
print("\n  Computing IG on full test set (848 instances)...")
X_test_t = torch.FloatTensor(X_test_sc.values.astype(np.float32))

ig_full, delta_full = ig.attribute(
    X_test_t,
    baseline_tensor,
    n_steps=50,
    return_convergence_delta=True
)
ig_full_np    = ig_full.detach().numpy()      # (848, 15)
delta_full_np = delta_full.detach().numpy()   # (848,)

# Completeness check: max |delta| should be small
max_delta = np.abs(delta_full_np).max()
mean_delta = np.abs(delta_full_np).mean()
print(f"\n  Completeness (convergence delta):")
print(f"    Max  |delta| : {max_delta:.6f}")
print(f"    Mean |delta| : {mean_delta:.6f}")
if max_delta < 0.05:
    print(f"    Status: GOOD — approximation error is small")
else:
    print(f"    Status: WARN — consider increasing n_steps")

# Global importance
ig_global = pd.Series(
    np.abs(ig_full_np).mean(axis=0),
    index=X_test_sc.columns
).sort_values(ascending=False)

print(f"\n  Top 5 features by mean |IG attribution|:")
for feat, val in ig_global.head(5).items():
    print(f"    {feat:<20} {val:.6f}")

# ── Stability instances IG ────────────────────────────────────────────────────
print("\n  Computing IG on 30 stability instances...")
X_stab_t = torch.FloatTensor(X_stability.values.astype(np.float32))

ig_stab, delta_stab = ig.attribute(
    X_stab_t,
    baseline_tensor,
    n_steps=50,
    return_convergence_delta=True
)
ig_stab_np = ig_stab.detach().numpy()    # (30, 15)

print(f"  Done. Shape: {ig_stab_np.shape}")

# ── Store results ─────────────────────────────────────────────────────────────
ig_store = {
    "global_importance" : ig_global,
    "full_test_ig"      : ig_full_np,    # (848, 15)
    "stability_ig"      : ig_stab_np,    # (30, 15)
    "baseline"          : baseline_np,
    "convergence_delta" : delta_full_np
}

# ── Four-way comparison: LR, RF, MLP-SHAP, MLP-IG ────────────────────────────
print("\n" + "=" * 60)
print("FOUR-WAY FEATURE RANKING COMPARISON")
print("=" * 60)
print(f"  {'Rank':<5} {'LR-SHAP':<20} {'RF-SHAP':<20} "
      f"{'MLP-SHAP':<20} {'MLP-IG':<20}")
print(f"  {'-'*4} {'-'*19} {'-'*19} {'-'*19} {'-'*19}")

lr_top  = shap_store["global_importance"]["LR"].index.tolist()
rf_top  = shap_store["global_importance"]["RF"].index.tolist()
mlp_top = shap_store["global_importance"]["MLP"].index.tolist()
ig_top  = ig_global.index.tolist()

for i in range(7):
    print(f"  {i+1:<5} {lr_top[i]:<20} {rf_top[i]:<20} "
          f"{mlp_top[i]:<20} {ig_top[i]:<20}")

# ── Jaccard: all methods vs LR anchor ─────────────────────────────────────────
def jaccard(a, b, k):
    s1, s2 = set(a[:k]), set(b[:k])
    return len(s1 & s2) / len(s1 | s2)

print(f"\n  Jaccard similarity vs LR-SHAP anchor:")
print(f"  {'Method':<12} {'Top-3':>8}  {'Top-5':>8}")
for name, ranking in [("RF-SHAP", rf_top),
                       ("MLP-SHAP", mlp_top),
                       ("MLP-IG", ig_top)]:
    j3 = jaccard(lr_top, ranking, 3)
    j5 = jaccard(lr_top, ranking, 5)
    s3 = "PASS" if j3 > 0.67 else "MARGINAL" if j3 > 0.33 else "FAIL"
    s5 = "PASS" if j5 > 0.67 else "MARGINAL" if j5 > 0.33 else "FAIL"
    print(f"  {name:<12} {j3:>8.4f} [{s3}]  {j5:>8.4f} [{s5}]")

# ── Clinical reference alignment preview ─────────────────────────────────────
print(f"\n  ACC/AHA Reference (top 7 for Framingham features):")
ref = ["age", "sysBP", "totChol", "currentSmoker",
       "glucose", "male", "heartRate"]
print(f"  {ref}")
print(f"\n  LR-SHAP top 7 : {lr_top[:7]}")
print(f"  RF-SHAP top 7 : {rf_top[:7]}")
print(f"  MLP-SHAP top 7: {mlp_top[:7]}")
print(f"  MLP-IG top 7  : {ig_top[:7]}")

print("\n  IG computation complete. ig_store populated.")
print("  Proceed to Cell 7 (Clinical Alignment — Kendall's Tau).")
```

```python
# Cell 7: Clinical Alignment — Kendall's Tau
# Reference: ACC/AHA 2019 Primary Prevention Guidelines
# Frozen before experiments, derived from Wilson 1998 + Arnett 2019
# Kendall's tau between model feature ranking and clinical reference ranking

import numpy as np
import pandas as pd
from scipy.stats import kendalltau
import warnings
warnings.filterwarnings("ignore")

print("=" * 60)
print("CELL 7 — CLINICAL ALIGNMENT (Kendall's Tau)")
print("=" * 60)

# ── Frozen clinical reference — ACC/AHA 2019 ──────────────────────────────────
# Lower rank number = more important clinically
# Features tied at same rank share that rank number
# Features not in reference assigned rank 8 (unranked — go to bottom)
# This ranking was frozen BEFORE experiments ran

CLINICAL_REFERENCE = {
    "age"           : 1,    # Rank 1 — strongest independent predictor
    "sysBP"         : 2,    # Rank 2 — Class I Level A, most modifiable
    "totChol"       : 3,    # Rank 3 — Class I Level A, LDL target
    "currentSmoker" : 4,    # Rank 4 — Class I Level B, doubles risk
    "cigsPerDay"    : 4,    # Rank 4 — same factor, more granular
    "glucose"       : 5,    # Rank 5 — Class IIa, ~2x risk
    "diabetes"      : 5,    # Rank 5 — same factor, binary form
    "male"          : 6,    # Rank 6 — component of Pooled Cohort Equations
    "heartRate"     : 7,    # Rank 7 — supporting evidence, smaller effect
    # All other features not in ACC/AHA upstream reference:
    # diaBP, BMI, education, prevalentHyp, prevalentStroke,
    # BPMeds — assigned rank 8 in get_clinical_rank()
}

UNRANKED_SCORE = 8   # assigned to features absent from reference

def get_clinical_rank(feature):
    return CLINICAL_REFERENCE.get(feature, UNRANKED_SCORE)

def kendalls_tau_alignment(importance_series, method_name, verbose=True):
    """
    Compute Kendall's tau between model feature ranking
    and ACC/AHA clinical reference ranking.

    importance_series: pd.Series, index=feature names,
                       values=mean |attribution|, sorted descending
    Returns tau, p_value
    """
    features = importance_series.index.tolist()
    n        = len(features)

    # Model rank: position in importance_series (1=most important)
    model_ranks    = list(range(1, n + 1))

    # Clinical rank: from frozen reference
    clinical_ranks = [get_clinical_rank(f) for f in features]

    tau, p_val = kendalltau(model_ranks, clinical_ranks)

    if verbose:
        print(f"\n  {method_name}")
        print(f"    Features evaluated : {n}")
        print(f"    Kendall's tau      : {tau:.4f}")
        print(f"    p-value            : {p_val:.4f}")

        # Threshold classification per PRD §7.3
        if tau >= 0.60:
            verdict = "PASS      (τ ≥ 0.60)"
        elif tau >= 0.40:
            verdict = "MARGINAL  (0.40 ≤ τ < 0.60)"
        else:
            verdict = "FAIL      (τ < 0.40)"
        print(f"    Verdict            : {verdict}")

        # Show feature-by-feature breakdown for transparency
        print(f"\n    Feature alignment breakdown:")
        print(f"    {'Feature':<20} {'Model Rank':>10}  "
              f"{'Clinical Rank':>13}  {'Direction'}")
        print(f"    {'-'*19} {'-'*10}  {'-'*13}  {'-'*12}")
        for feat, mr, cr in zip(features, model_ranks, clinical_ranks):
            direction = "↑ over-ranked" if mr < cr else \
                        "↓ under-ranked" if mr > cr else "= aligned"
            flag = " ←" if feat not in CLINICAL_REFERENCE else ""
            print(f"    {feat:<20} {mr:>10}  {cr:>13}  {direction}{flag}")

    return tau, p_val

# ── Run for all four method-model combinations ────────────────────────────────
print("\n── Framingham Clinical Alignment Results ──")
print("   Reference: ACC/AHA 2019 + Framingham Risk Score")
print("   ← marks features absent from clinical reference (rank=8)\n")

tau_results = {}

# LR-SHAP
tau, p = kendalls_tau_alignment(
    shap_store["global_importance"]["LR"], "LR-SHAP"
)
tau_results["LR_SHAP"] = {"tau": tau, "p": p}

# RF-SHAP
tau, p = kendalls_tau_alignment(
    shap_store["global_importance"]["RF"], "RF-SHAP"
)
tau_results["RF_SHAP"] = {"tau": tau, "p": p}

# MLP-SHAP
tau, p = kendalls_tau_alignment(
    shap_store["global_importance"]["MLP"], "MLP-SHAP"
)
tau_results["MLP_SHAP"] = {"tau": tau, "p": p}

# MLP-IG
tau, p = kendalls_tau_alignment(
    ig_store["global_importance"], "MLP-IG"
)
tau_results["MLP_IG"] = {"tau": tau, "p": p}

# ── Summary table ─────────────────────────────────────────────────────────────
print("\n" + "=" * 60)
print("ALIGNMENT SUMMARY — FRAMINGHAM")
print("=" * 60)
print(f"  {'Method':<14} {'Tau':>8}  {'p-val':>8}  {'Verdict'}")
print(f"  {'-'*13} {'-'*8}  {'-'*8}  {'-'*20}")
for method, res in tau_results.items():
    tau, p = res["tau"], res["p"]
    if tau >= 0.60:
        v = "PASS"
    elif tau >= 0.40:
        v = "MARGINAL"
    else:
        v = "FAIL"
    sig = "p<0.05" if p < 0.05 else "n.s."
    print(f"  {method:<14} {tau:>8.4f}  {p:>8.4f}  {v} ({sig})")

# ── Key features driving misalignment ────────────────────────────────────────
print("\n── Misalignment Analysis ──")
print("""
  Features consistently over-ranked (appear higher in model than ACC/AHA):
    - prevalentHyp  : consequence of high BP, not upstream risk factor
    - diaBP         : correlated with sysBP (r~0.78), attribution split
    - education     : socioeconomic proxy, not direct biological mechanism
    - male          : LR places at rank 2; ACC/AHA rank 6

  Features consistently under-ranked (appear lower in model than ACC/AHA):
    - totChol       : LDL target in guidelines, but Framingham data has
                      high measurement noise and outliers (max=696 mg/dL)
    - glucose       : 9.15% missing, imputed values dilute signal
    - heartRate     : genuinely smaller effect, consistently underweighted

  These patterns are expected and interpretable:
    - Collinear BP features split attribution (sysBP + diaBP + prevalentHyp)
    - Missing data imputation dilutes glucose signal
    - Education acting as confounder in RF specifically
""")

print("  Proceed to Cell 8 (Stability Evaluation).")
```

```python
# Cell 8: Stability Evaluation
# 30 fixed test instances × 50 perturbations × 3 noise levels
# Noise applied to continuous features only
# Metric: mean rank perturbation at 2% noise (primary reporting level)
# Threshold: <1.0 positions = PASS, 1.0-2.0 = MARGINAL, >2.0 = FAIL

import numpy as np
import pandas as pd
import shap
import torch
from captum.attr import IntegratedGradients
import warnings
warnings.filterwarnings("ignore")

RANDOM_SEED  = 42
N_PERTURB    = 50
NOISE_LEVELS = [0.01, 0.02, 0.05]

print("=" * 60)
print("CELL 8 — STABILITY EVALUATION")
print(f"  {len(X_stability)} instances × {N_PERTURB} perturbations "
      f"× {len(NOISE_LEVELS)} noise levels")
print("=" * 60)

# ── Feature type split ────────────────────────────────────────────────────────
# Only continuous features get perturbed
# Categorical features are binary/ordinal — Gaussian noise is meaningless
CATEGORICAL_FEATS = ["male", "education", "currentSmoker", "BPMeds",
                     "prevalentStroke", "prevalentHyp", "diabetes"]
CONTINUOUS_FEATS  = [c for c in X_test_sc.columns
                     if c not in CATEGORICAL_FEATS]

print(f"\n  Continuous features perturbed ({len(CONTINUOUS_FEATS)}): "
      f"{CONTINUOUS_FEATS}")
print(f"  Categorical features held fixed ({len(CATEGORICAL_FEATS)}): "
      f"{CATEGORICAL_FEATS}")

# ── Feature ranges from ORIGINAL training data ────────────────────────────────
# Range computed in scaled space (what the models see)
feature_ranges = X_train_sc[CONTINUOUS_FEATS].max() - \
                 X_train_sc[CONTINUOUS_FEATS].min()

print(f"\n  Feature ranges (scaled space) used for noise scaling:")
for feat, rng in feature_ranges.items():
    print(f"    {feat:<20} range={rng:.4f}")

# ── Rebuild explainers ────────────────────────────────────────────────────────
print("\n  Rebuilding explainers...")

lr_model  = final_results["LR"]["model"]
rf_model  = final_results["RF"]["model"]
mlp_model = final_results["MLP"]["model"]
mlp_model.eval()

explainer_lr  = shap.LinearExplainer(lr_model, X_train_sc)
explainer_rf  = shap.TreeExplainer(rf_model)

background    = shap.sample(X_train_sc, 100, random_state=RANDOM_SEED)
def mlp_predict(x):
    x_t = torch.FloatTensor(x)
    with torch.no_grad():
        return torch.sigmoid(mlp_model(x_t)).numpy().flatten()
explainer_mlp = shap.KernelExplainer(mlp_predict, background)

class MLPWrapper(torch.nn.Module):
    def __init__(self, m): super().__init__(); self.model = m
    def forward(self, x): return torch.sigmoid(self.model(x))

wrapper         = MLPWrapper(mlp_model)
ig_explainer    = IntegratedGradients(wrapper)
baseline_tensor = torch.FloatTensor(
    X_train_sc.mean(axis=0).values.astype(np.float32)
).unsqueeze(0)

print("  Explainers ready.")

# ── Rank utility ──────────────────────────────────────────────────────────────
def to_ranks(attr_matrix):
    """
    attr_matrix: (n_instances, n_features) numpy array
    Returns rank matrix where rank 1 = highest |attribution|
    """
    ranks = np.zeros_like(attr_matrix, dtype=float)
    for i, row in enumerate(attr_matrix):
        order = np.argsort(np.abs(row))[::-1]
        for rank_pos, feat_idx in enumerate(order):
            ranks[i, feat_idx] = rank_pos + 1
    return ranks

# ── Pre-compute baseline ranks for 30 stability instances ─────────────────────
print("\n  Computing baseline ranks for 30 stability instances...")

base_shap_lr  = explainer_lr.shap_values(X_stability)

base_shap_rf_raw = explainer_rf.shap_values(X_stability)
if isinstance(base_shap_rf_raw, list):
    base_shap_rf = base_shap_rf_raw[1]
elif base_shap_rf_raw.ndim == 3:
    base_shap_rf = base_shap_rf_raw[:, :, 1]
else:
    base_shap_rf = base_shap_rf_raw

base_shap_mlp = shap_store["stability_shap"]["MLP"]   # already computed
base_ig       = ig_store["stability_ig"]               # already computed

base_ranks = {
    "LR_SHAP" : to_ranks(base_shap_lr),
    "RF_SHAP" : to_ranks(base_shap_rf),
    "MLP_SHAP": to_ranks(np.array(base_shap_mlp)),
    "MLP_IG"  : to_ranks(base_ig)
}
print("  Baseline ranks computed.")

# ── Main perturbation loop ────────────────────────────────────────────────────
# Storage: results[method][noise_level] = list of mean rank shifts
#          one value per stability instance
results_stability = {m: {nl: [] for nl in NOISE_LEVELS}
                     for m in ["LR_SHAP", "RF_SHAP", "MLP_SHAP", "MLP_IG"]}

np.random.seed(RANDOM_SEED)

print(f"\n  Running perturbation loop...")
print(f"  {'Instance':>10}  {'Noise':>6}  "
      f"{'LR_SHAP':>9}  {'RF_SHAP':>9}  "
      f"{'MLP_SHAP':>9}  {'MLP_IG':>9}")
print(f"  {'-'*70}")

for nl in NOISE_LEVELS:
    print(f"\n  Noise level: {nl*100:.0f}%")

    for i in range(len(X_stability)):
        instance = X_stability.iloc[i:i+1].copy()

        # Generate 50 perturbed versions
        perturbed = pd.concat([instance] * N_PERTURB, ignore_index=True)
        for feat in CONTINUOUS_FEATS:
            std = nl * feature_ranges[feat]
            perturbed[feat] += np.random.normal(0, std, N_PERTURB)

        # LR SHAP
        sh_lr    = explainer_lr.shap_values(perturbed)
        ranks_lr = to_ranks(sh_lr)
        shift_lr = np.mean(np.abs(ranks_lr - base_ranks["LR_SHAP"][i]))
        results_stability["LR_SHAP"][nl].append(shift_lr)

        # RF SHAP
        sh_rf_raw = explainer_rf.shap_values(perturbed)
        if isinstance(sh_rf_raw, list):
            sh_rf = sh_rf_raw[1]
        elif sh_rf_raw.ndim == 3:
            sh_rf = sh_rf_raw[:, :, 1]
        else:
            sh_rf = sh_rf_raw
        ranks_rf = to_ranks(sh_rf)
        shift_rf = np.mean(np.abs(ranks_rf - base_ranks["RF_SHAP"][i]))
        results_stability["RF_SHAP"][nl].append(shift_rf)

        # MLP SHAP
        sh_mlp    = explainer_mlp.shap_values(perturbed, silent=True)
        ranks_mlp = to_ranks(np.array(sh_mlp))
        shift_mlp = np.mean(np.abs(ranks_mlp - base_ranks["MLP_SHAP"][i]))
        results_stability["MLP_SHAP"][nl].append(shift_mlp)

        # MLP IG
        pert_t   = torch.FloatTensor(perturbed.values.astype(np.float32))
        ig_pert  = ig_explainer.attribute(
            pert_t, baseline_tensor, n_steps=50
        ).detach().numpy()
        ranks_ig = to_ranks(ig_pert)
        shift_ig = np.mean(np.abs(ranks_ig - base_ranks["MLP_IG"][i]))
        results_stability["MLP_IG"][nl].append(shift_ig)

        if (i + 1) % 10 == 0:
            print(f"  Instance {i+1:>3}/30  noise={nl*100:.0f}%  "
                  f"LR={shift_lr:.2f}  RF={shift_rf:.2f}  "
                  f"MLP={shift_mlp:.2f}  IG={shift_ig:.2f}")

# ── Aggregate results ─────────────────────────────────────────────────────────
print("\n\n" + "=" * 60)
print("STABILITY RESULTS — MEAN RANK PERTURBATION")
print("=" * 60)
print(f"\n  {'Method':<12} "
      f"{'1% noise':>10}  {'2% noise':>10}  {'5% noise':>10}  "
      f"{'Verdict@2%':>12}")
print(f"  {'-'*11} {'-'*10}  {'-'*10}  {'-'*10}  {'-'*12}")

stability_summary = {}
for method in ["LR_SHAP", "RF_SHAP", "MLP_SHAP", "MLP_IG"]:
    vals = {}
    for nl in NOISE_LEVELS:
        vals[nl] = np.mean(results_stability[method][nl])

    v2 = vals[0.02]
    if v2 < 1.0:
        verdict = "PASS"
    elif v2 <= 2.0:
        verdict = "MARGINAL"
    else:
        verdict = "FAIL"

    stability_summary[method] = {"values": vals, "verdict": verdict}

    print(f"  {method:<12} "
          f"{vals[0.01]:>10.4f}  {vals[0.02]:>10.4f}  {vals[0.05]:>10.4f}  "
          f"{verdict:>12}")

# ── Clinical noise context ────────────────────────────────────────────────────
print(f"""
  Clinical context:
    Blood pressure measurement variability in clinical practice:
    3-8 mmHg between readings in the same session.
    In this scaled dataset, 2% of sysBP range ≈ this variability.
    A method that fails at 2% noise is unreliable at the precision
    level of a standard clinical measurement.
""")

# Store for reliability map
stability_store = stability_summary
print("  stability_store populated. Proceed to Cell 9 (Consistency).")
```

```python
# Cell 9: Consistency Evaluation
# Sub-A: Cross-seed consistency — 10 seeds for RF and MLP
#        Spearman rank correlation between global importance rankings
#        Key comparison: AUC variance vs explanation variance
# Sub-B: Cross-model Jaccard (formalised from Cell 6 results)

import numpy as np
import pandas as pd
import shap
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from scipy.stats import spearmanr
import copy
import warnings
warnings.filterwarnings("ignore")

RANDOM_SEED = 42
SEEDS       = list(range(10))

print("=" * 60)
print("CELL 9 — CONSISTENCY EVALUATION")
print("=" * 60)

# ── Background for KernelExplainer (fixed across seeds) ──────────────────────
background = shap.sample(X_train_sc, 100, random_state=RANDOM_SEED)

# ── Storage ───────────────────────────────────────────────────────────────────
rf_aucs           = []
mlp_aucs          = []
rf_shap_rankings  = []   # one pd.Series per seed
mlp_shap_rankings = []
mlp_ig_rankings   = []

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ── Sub-Experiment A: Cross-seed retraining ───────────────────────────────────
print("\n── Sub-Experiment A: Cross-seed consistency (10 seeds) ──")
print(f"  RF : n_estimators=500, max_depth=8, min_samples_leaf=15")
print(f"  MLP: same architecture, 50 epochs (speed), same SMOTE train data")
print(f"  Explaining on subset of 100 test instances for speed\n")

# Subset of test instances for SHAP in consistency loop
# Using 100 instances — enough for stable global importance ranking
# without 10x the KernelExplainer runtime
np.random.seed(RANDOM_SEED)
consistency_idx = np.random.choice(len(X_test_sc), size=100, replace=False)
X_consist       = X_test_sc.iloc[consistency_idx]
y_consist       = y_test.iloc[consistency_idx]
X_consist_t     = torch.FloatTensor(X_consist.values.astype(np.float32))

baseline_tensor = torch.FloatTensor(
    X_train_sc.mean(axis=0).values.astype(np.float32)
).unsqueeze(0)

print(f"  {'Seed':>5}  {'RF AUC':>8}  {'MLP AUC':>9}  Status")
print(f"  {'-'*5}  {'-'*8}  {'-'*9}  {'-'*20}")

for seed in SEEDS:
    # ── RF ────────────────────────────────────────────────────────────────────
    rf_s = RandomForestClassifier(
        n_estimators=500, max_depth=8,
        min_samples_leaf=15, max_features="sqrt",
        random_state=seed, n_jobs=-1
    )
    rf_s.fit(X_train_bal, y_train_bal)
    rf_prob_s = rf_s.predict_proba(X_test_sc)[:, 1]
    rf_auc_s  = roc_auc_score(y_test, rf_prob_s)
    rf_aucs.append(rf_auc_s)

    # RF SHAP global importance
    exp_rf_s     = shap.TreeExplainer(rf_s)
    sh_rf_s_raw  = exp_rf_s.shap_values(X_consist)
    if isinstance(sh_rf_s_raw, list):
        sh_rf_s = sh_rf_s_raw[1]
    elif sh_rf_s_raw.ndim == 3:
        sh_rf_s = sh_rf_s_raw[:, :, 1]
    else:
        sh_rf_s = sh_rf_s_raw
    rf_imp_s = pd.Series(
        np.abs(sh_rf_s).mean(axis=0), index=X_consist.columns
    )
    rf_shap_rankings.append(rf_imp_s)

    # ── MLP ───────────────────────────────────────────────────────────────────
    torch.manual_seed(seed)
    np.random.seed(seed)

    class ClinicalMLP(nn.Module):
        def __init__(self, d):
            super().__init__()
            self.net = nn.Sequential(
                nn.Linear(d, 64), nn.ReLU(), nn.Dropout(0.2),
                nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2),
                nn.Linear(32, 1)
            )
        def forward(self, x): return self.net(x)

    mlp_s     = ClinicalMLP(X_train_bal.shape[1]).to(device)
    criterion = nn.BCEWithLogitsLoss(
        pos_weight=torch.tensor([1.0]).to(device)
    )
    optimizer = optim.Adam(mlp_s.parameters(), lr=0.0005,
                            weight_decay=5e-3)

    X_tr_t = torch.FloatTensor(X_train_bal.values.astype(np.float32)).to(device)
    y_tr_t = torch.FloatTensor(y_train_bal.values.astype(np.float32)).unsqueeze(1).to(device)
    loader = DataLoader(TensorDataset(X_tr_t, y_tr_t),
                        batch_size=32, shuffle=True)

    # 50 epochs — enough for ranking stability check, faster than full training
    mlp_s.train()
    for epoch in range(50):
        for bX, by in loader:
            optimizer.zero_grad()
            criterion(mlp_s(bX), by).backward()
            optimizer.step()

    mlp_s.eval()
    X_test_t_gpu = torch.FloatTensor(
        X_test_sc.values.astype(np.float32)
    ).to(device)
    with torch.no_grad():
        mlp_prob_s = torch.sigmoid(
            mlp_s(X_test_t_gpu)
        ).cpu().numpy().flatten()
    mlp_auc_s = roc_auc_score(y_test, mlp_prob_s)
    mlp_aucs.append(mlp_auc_s)
    mlp_s_cpu = mlp_s.to("cpu")

    # MLP SHAP global importance
    def mlp_predict_s(x):
        x_t = torch.FloatTensor(x)
        with torch.no_grad():
            return torch.sigmoid(mlp_s_cpu(x_t)).numpy().flatten()

    exp_mlp_s = shap.KernelExplainer(mlp_predict_s, background)
    sh_mlp_s  = exp_mlp_s.shap_values(X_consist, silent=True)
    mlp_imp_s = pd.Series(
        np.abs(np.array(sh_mlp_s)).mean(axis=0),
        index=X_consist.columns
    )
    mlp_shap_rankings.append(mlp_imp_s)

    # MLP IG global importance
    class MLPWrapper(nn.Module):
        def __init__(self, m): super().__init__(); self.model = m
        def forward(self, x): return torch.sigmoid(self.model(x))

    from captum.attr import IntegratedGradients
    ig_s     = IntegratedGradients(MLPWrapper(mlp_s_cpu))
    ig_attr_s = ig_s.attribute(
        X_consist_t, baseline_tensor, n_steps=50
    ).detach().numpy()
    mlp_ig_s = pd.Series(
        np.abs(ig_attr_s).mean(axis=0), index=X_consist.columns
    )
    mlp_ig_rankings.append(mlp_ig_s)

    print(f"  {seed:>5}  {rf_auc_s:>8.4f}  {mlp_auc_s:>9.4f}  done")

# ── Spearman rank correlations ────────────────────────────────────────────────
def pairwise_spearman(rankings_list):
    """
    Compute all 45 pairwise Spearman correlations across 10 seeds.
    Returns mean, std, min.
    """
    corrs = []
    n     = len(rankings_list)
    for i in range(n):
        for j in range(i + 1, n):
            r1 = rankings_list[i].rank(ascending=False).values
            r2 = rankings_list[j].rank(ascending=False).values
            rho, _ = spearmanr(r1, r2)
            corrs.append(rho)
    return np.mean(corrs), np.std(corrs), np.min(corrs), corrs

rf_spear  = pairwise_spearman(rf_shap_rankings)
mlp_spear = pairwise_spearman(mlp_shap_rankings)
ig_spear  = pairwise_spearman(mlp_ig_rankings)

print("\n" + "=" * 60)
print("SUB-EXPERIMENT A RESULTS")
print("=" * 60)

print(f"\n  Prediction variance across 10 seeds:")
print(f"    RF  AUC — mean={np.mean(rf_aucs):.4f}  "
      f"std={np.std(rf_aucs):.5f}  "
      f"range=[{min(rf_aucs):.4f}, {max(rf_aucs):.4f}]")
print(f"    MLP AUC — mean={np.mean(mlp_aucs):.4f}  "
      f"std={np.std(mlp_aucs):.5f}  "
      f"range=[{min(mlp_aucs):.4f}, {max(mlp_aucs):.4f}]")

print(f"\n  Explanation variance across 10 seeds (Spearman rho):")
print(f"  {'Method':<12} {'Mean':>8}  {'Std':>8}  {'Min':>8}  Verdict")
print(f"  {'-'*11} {'-'*8}  {'-'*8}  {'-'*8}  {'-'*12}")

for name, (mean, std, mn, _) in [
    ("RF_SHAP",  rf_spear),
    ("MLP_SHAP", mlp_spear),
    ("MLP_IG",   ig_spear)
]:
    v = "PASS" if mean > 0.75 else "MARGINAL" if mean >= 0.55 else "FAIL"
    print(f"  {name:<12} {mean:>8.4f}  {std:>8.4f}  {mn:>8.4f}  {v}")

# ── Key comparison: prediction stable, explanation stable? ────────────────────
print(f"\n  Key comparison (PRD §7.2 primary finding):")
rf_auc_var  = np.var(rf_aucs)
mlp_auc_var = np.var(mlp_aucs)
print(f"    RF  — AUC variance={rf_auc_var:.6f} | "
      f"SHAP Spearman mean={rf_spear[0]:.4f}")
print(f"    MLP — AUC variance={mlp_auc_var:.6f} | "
      f"SHAP Spearman mean={mlp_spear[0]:.4f} | "
      f"IG Spearman mean={ig_spear[0]:.4f}")

if mlp_auc_var < 0.001 and ig_spear[0] < 0.75:
    print(f"\n    *** FINDING: MLP predictions are stable across seeds")
    print(f"        (AUC variance={mlp_auc_var:.6f}) but explanation")
    print(f"        rankings are not (IG Spearman={ig_spear[0]:.4f}).")
    print(f"        Same model, same data, different initialization →")
    print(f"        different explanations despite identical predictions.")

# ── Sub-Experiment B: Cross-model Jaccard (formal) ────────────────────────────
print("\n" + "=" * 60)
print("SUB-EXPERIMENT B — CROSS-MODEL JACCARD (seed=42 models)")
print("=" * 60)

def jaccard(a, b, k):
    s1, s2 = set(a[:k]), set(b[:k])
    return len(s1 & s2) / len(s1 | s2)

lr_top  = shap_store["global_importance"]["LR"].index.tolist()
rf_top  = shap_store["global_importance"]["RF"].index.tolist()
mlp_top = shap_store["global_importance"]["MLP"].index.tolist()
ig_top  = ig_store["global_importance"].index.tolist()

print(f"\n  {'Method':<12}  {'Top-3':>8}  {'Verdict':>10}  "
      f"{'Top-5':>8}  {'Verdict':>10}")
print(f"  {'-'*11}  {'-'*8}  {'-'*10}  {'-'*8}  {'-'*10}")

consistency_store = {
    "rf_shap" : rf_spear,
    "mlp_shap": mlp_spear,
    "mlp_ig"  : ig_spear,
    "rf_aucs" : rf_aucs,
    "mlp_aucs": mlp_aucs
}

jaccard_store = {}
for name, ranking in [("RF_SHAP", rf_top),
                       ("MLP_SHAP", mlp_top),
                       ("MLP_IG", ig_top)]:
    j3 = jaccard(lr_top, ranking, 3)
    j5 = jaccard(lr_top, ranking, 5)
    v3 = "PASS" if j3 > 0.67 else "MARGINAL" if j3 > 0.33 else "FAIL"
    v5 = "PASS" if j5 > 0.67 else "MARGINAL" if j5 > 0.33 else "FAIL"
    jaccard_store[name] = {"top3": j3, "top5": j5}
    print(f"  {name:<12}  {j3:>8.4f}  {v3:>10}  {j5:>8.4f}  {v5:>10}")

print("\n  consistency_store and jaccard_store populated.")
print("  Proceed to Cell 10 (Reliability Map).")
```

```python
# Cell 13: BRFSS 2015 Preprocessing
# No imputation needed (zero missing values)
# SMOTE on training only, test set untouched
# Stratified 80/20 split, seed=42

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

RANDOM_SEED  = 42
TARGET_BRFSS = "HeartDiseaseorAttack"

print("=" * 55)
print("BRFSS 2015 — PREPROCESSING")
print("=" * 55)

# ── Feature decisions — documented for paper ──────────────────────────────────
print("""
  Feature decisions (documented for paper):
    Age      : ordinal bands 1–13 (18-24 to 80+)
               treated as ordinal continuous — monotonic with risk
    Diabetes : ordinal 0/1/2 (no/pre/diabetic)
               treated as ordinal continuous — ordering clinically meaningful
    All others: binary (0/1) or continuous (BMI, MentHlth, PhysHlth)
    No imputation required — zero missing values
""")

# ── All features except target ────────────────────────────────────────────────
FEATURE_COLS = [c for c in df_brfss.columns if c != TARGET_BRFSS]

X = df_brfss[FEATURE_COLS].copy()
y = df_brfss[TARGET_BRFSS].copy().astype(int)

print(f"  Features : {len(FEATURE_COLS)}")
print(f"  Samples  : {len(X):,}")

# ── Stratified 80/20 split ────────────────────────────────────────────────────
X_train_b_raw, X_test_b_raw, y_train_b, y_test_b = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=RANDOM_SEED
)

print(f"\n  Train raw : {len(X_train_b_raw):,} rows")
print(f"  Test  raw : {len(X_test_b_raw):,} rows")
print(f"\n  Train class distribution:")
print(f"    Class 0 : {(y_train_b==0).sum():,}  ({(y_train_b==0).mean()*100:.1f}%)")
print(f"    Class 1 : {(y_train_b==1).sum():,}  ({(y_train_b==1).mean()*100:.1f}%)")

# ── StandardScaler — fitted on train only ─────────────────────────────────────
# All 21 features scaled — no imputation step needed
scaler_b   = StandardScaler()
X_train_b_sc = pd.DataFrame(
    scaler_b.fit_transform(X_train_b_raw),
    columns=FEATURE_COLS, index=X_train_b_raw.index
)
X_test_b_sc  = pd.DataFrame(
    scaler_b.transform(X_test_b_raw),
    columns=FEATURE_COLS, index=X_test_b_raw.index
)
print(f"\n  StandardScaler fitted on train, applied to test.")

# ── SMOTE — train only ────────────────────────────────────────────────────────
smote_b = SMOTE(random_state=RANDOM_SEED, k_neighbors=5)
X_train_b_bal_arr, y_train_b_bal_arr = smote_b.fit_resample(
    X_train_b_sc, y_train_b
)
X_train_b_bal = pd.DataFrame(X_train_b_bal_arr, columns=FEATURE_COLS)
y_train_b_bal = pd.Series(y_train_b_bal_arr, name=TARGET_BRFSS)

print(f"\n  SMOTE applied to training set only:")
print(f"    Before — Class 0: {(y_train_b==0).sum():,}  "
      f"Class 1: {(y_train_b==1).sum():,}  "
      f"Total: {len(y_train_b):,}")
print(f"    After  — Class 0: {(y_train_b_bal==0).sum():,}  "
      f"Class 1: {(y_train_b_bal==1).sum():,}  "
      f"Total: {len(y_train_b_bal):,}")
print(f"    Synthetic samples added: "
      f"{len(y_train_b_bal) - len(y_train_b):,}")

# ── Stratified explanation subsample from test set ────────────────────────────
# KernelExplainer on 50k+ instances is infeasible
# 500 stratified instances give stable global importance rankings
np.random.seed(RANDOM_SEED)

idx_0 = X_test_b_sc[y_test_b == 0].index
idx_1 = X_test_b_sc[y_test_b == 1].index

# Sample proportional to test set class distribution (~91/9)
n_explain   = 500
n_explain_1 = max(int(n_explain * (y_test_b==1).mean()), 20)
n_explain_0 = n_explain - n_explain_1

sampled_0 = np.random.choice(idx_0, n_explain_0, replace=False)
sampled_1 = np.random.choice(idx_1, n_explain_1, replace=False)
explain_idx = np.concatenate([sampled_0, sampled_1])
np.random.shuffle(explain_idx)

X_explain_b = X_test_b_sc.loc[explain_idx]
y_explain_b = y_test_b.loc[explain_idx]

print(f"\n  Explanation subsample (stratified from test set):")
print(f"    Total  : {len(X_explain_b)}")
print(f"    Class 0: {(y_explain_b==0).sum()}  "
      f"Class 1: {(y_explain_b==1).sum()}")

# ── Final shape summary ───────────────────────────────────────────────────────
print(f"\n  Final shapes:")
print(f"    X_train_b_bal : {X_train_b_bal.shape}  "
      f"(SMOTE-balanced, model training)")
print(f"    X_test_b_sc   : {X_test_b_sc.shape}   "
      f"(original patients, evaluation)")
print(f"    X_explain_b   : {X_explain_b.shape}   "
      f"(stratified subsample for SHAP/IG)")
print(f"    y_test_b      : {y_test_b.shape}")

# ── Freeze BRFSS clinical reference ───────────────────────────────────────────
# Must be documented as frozen before models run
CLINICAL_REF_BRFSS = {
    "Age"    : 1,   # Rank 1 — strongest predictor (ordinal bands)
    "HighBP" : 2,   # Rank 2 — binary proxy for systolic BP
    "HighChol": 3,  # Rank 3 — binary proxy for total cholesterol
    "Smoker" : 4,   # Rank 4 — smoking status
    "Diabetes": 5,  # Rank 5 — diabetes/pre-diabetes (ordinal 0/1/2)
    "Sex"    : 6,   # Rank 6 — component of Pooled Cohort Equations
    # No heart rate equivalent in BRFSS survey
    # All other features: unranked (lifestyle, access, general health)
}
UNRANKED_BRFSS = 7   # lower than Framingham's 8 — fewer ranked features

print(f"\n  BRFSS Clinical Reference (ACC/AHA mapped, frozen):")
for feat, rank in CLINICAL_REF_BRFSS.items():
    print(f"    Rank {rank}: {feat}")
print(f"    Rank {UNRANKED_BRFSS}: all other features "
      f"({len(FEATURE_COLS) - len(CLINICAL_REF_BRFSS)} features)")

print(f"""
  Limitation noted for paper:
    HighBP and HighChol are binary survey self-reports, not
    continuous clinical measurements (mmHg, mg/dL).
    This reduces attribution granularity compared to Framingham.
    Age is ordinal bands, not continuous years.
    These differences in feature granularity between datasets
    are themselves a finding — alignment may differ not because
    SHAP behaves differently but because the features represent
    the same constructs at different resolutions.
""")

print("  Preprocessing complete. Proceed to Cell 14 (BRFSS model training).")
```

```python
# Cell 14: BRFSS Model Training — LR, RF, MLP
# GridSearchCV on 50k stratified subsample (speed)
# Final models trained on full 367k SMOTE-balanced dataset
# Evaluated on original 50k test set (real patient distribution)

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import (accuracy_score, roc_auc_score, f1_score,
                              recall_score, precision_score,
                              confusion_matrix, roc_curve)
import numpy as np
import pandas as pd
import copy

RANDOM_SEED = 42

def youden_threshold(y_true, y_prob):
    fpr, tpr, thresholds = roc_curve(y_true, y_prob)
    j = tpr - fpr
    return float(np.clip(thresholds[np.argmax(j)], 0.10, 0.90))

def evaluate(name, y_true, y_prob, threshold):
    y_pred         = (y_prob >= threshold).astype(int)
    cm             = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    auc  = roc_auc_score(y_true, y_prob)
    sens = recall_score(y_true, y_pred)
    spec = tn / (tn + fp)
    print(f"\n  {name}")
    print(f"    AUC        : {auc:.4f}")
    print(f"    F1         : {f1_score(y_true, y_pred):.4f}")
    print(f"    Accuracy   : {accuracy_score(y_true, y_pred):.4f}")
    print(f"    Sensitivity: {sens:.4f}   ← critical")
    print(f"    Specificity: {spec:.4f}")
    print(f"    Precision  : {precision_score(y_true, y_pred):.4f}")
    print(f"    Threshold  : {threshold:.3f}")
    print(f"    CM → TN:{tn}  FP:{fp}  FN:{fn}  TP:{tp}")
    return auc, sens

print("=" * 60)
print("BRFSS MODEL TRAINING")
print("Train: SMOTE-balanced 367k  |  Test: original 50k")
print("=" * 60)

# ── Subsample for GridSearchCV (speed) ───────────────────────────────────────
# 50k stratified subsample from SMOTE-balanced training data
# Hyperparameters found here, final model trained on full dataset
print("\n  Creating 50k stratified subsample for GridSearchCV...")
sub_idx = train_test_split(
    np.arange(len(X_train_b_bal)),
    test_size=(1 - 50000/len(X_train_b_bal)),
    stratify=y_train_b_bal,
    random_state=RANDOM_SEED
)[0]
X_cv = X_train_b_bal.iloc[sub_idx]
y_cv = y_train_b_bal.iloc[sub_idx]
print(f"  CV subsample: {len(X_cv):,} rows  "
      f"(Class 0: {(y_cv==0).sum():,}  Class 1: {(y_cv==1).sum():,})")

# Validation split from full SMOTE train (threshold calibration)
X_tr_b, X_val_b, y_tr_b, y_val_b = train_test_split(
    X_train_b_bal, y_train_b_bal,
    test_size=0.10, stratify=y_train_b_bal,
    random_state=RANDOM_SEED
)
print(f"  Threshold calibration val set: {len(X_val_b):,} rows")

final_results_b = {}

# ═══════════════════════════════════════════════════════
# 1. LOGISTIC REGRESSION
# ═══════════════════════════════════════════════════════
print("\n── Logistic Regression ──")
lr_base_b = LogisticRegression(
    penalty="l2", solver="lbfgs", max_iter=2000,
    random_state=RANDOM_SEED, class_weight="balanced"
)
lr_gs_b = GridSearchCV(
    lr_base_b,
    {"C": [0.001, 0.01, 0.1, 1, 10]},
    cv=3, scoring="roc_auc", n_jobs=-1
)
lr_gs_b.fit(X_cv, y_cv)
best_C_b = lr_gs_b.best_params_["C"]
print(f"  Best C (from CV subsample): {best_C_b}")

# Retrain on full SMOTE dataset with best C
best_lr_b = LogisticRegression(
    C=best_C_b, penalty="l2", solver="lbfgs",
    max_iter=2000, random_state=RANDOM_SEED,
    class_weight="balanced"
)
best_lr_b.fit(X_train_b_bal, y_train_b_bal)

lr_thresh_b = youden_threshold(y_val_b,
                                best_lr_b.predict_proba(X_val_b)[:, 1])
lr_probs_b  = best_lr_b.predict_proba(X_test_b_sc)[:, 1]
lr_auc_b, lr_sens_b = evaluate("LR — Test Set", y_test_b,
                                 lr_probs_b, lr_thresh_b)
final_results_b["LR"] = {
    "model"    : best_lr_b,
    "threshold": lr_thresh_b,
    "auc"      : lr_auc_b,
    "sens"     : lr_sens_b,
    "probs"    : lr_probs_b
}

# ═══════════════════════════════════════════════════════
# 2. RANDOM FOREST
# ═══════════════════════════════════════════════════════
print("\n── Random Forest ──")
rf_base_b = RandomForestClassifier(
    n_estimators=300,
    max_features="sqrt",
    min_samples_leaf=15,
    random_state=RANDOM_SEED,
    n_jobs=-1
)
rf_gs_b = GridSearchCV(
    rf_base_b,
    {"max_depth": [3, 5, 8]},
    cv=3, scoring="roc_auc", n_jobs=-1
)
rf_gs_b.fit(X_cv, y_cv)
best_depth_b = rf_gs_b.best_params_["max_depth"]
print(f"  Best max_depth (from CV subsample): {best_depth_b}")

# CV score breakdown
for depth, mean_s, std_s in zip(
    rf_gs_b.cv_results_["param_max_depth"].data,
    rf_gs_b.cv_results_["mean_test_score"],
    rf_gs_b.cv_results_["std_test_score"]
):
    print(f"    depth={depth}  CV AUC={mean_s:.4f} ± {std_s:.4f}")

# Retrain on full SMOTE dataset
best_rf_b = RandomForestClassifier(
    n_estimators=300,
    max_depth=best_depth_b,
    max_features="sqrt",
    min_samples_leaf=15,
    random_state=RANDOM_SEED,
    n_jobs=-1
)
best_rf_b.fit(X_train_b_bal, y_train_b_bal)

rf_thresh_b = youden_threshold(y_val_b,
                                best_rf_b.predict_proba(X_val_b)[:, 1])
rf_probs_b  = best_rf_b.predict_proba(X_test_b_sc)[:, 1]
rf_auc_b, rf_sens_b = evaluate("RF — Test Set", y_test_b,
                                 rf_probs_b, rf_thresh_b)
final_results_b["RF"] = {
    "model"    : best_rf_b,
    "threshold": rf_thresh_b,
    "auc"      : rf_auc_b,
    "sens"     : rf_sens_b,
    "probs"    : rf_probs_b
}

# ═══════════════════════════════════════════════════════
# 3. MLP — same fixed architecture
#    Batch size 256 (larger dataset, stable gradients)
#    pos_weight=1.0 (SMOTE-balanced training)
# ═══════════════════════════════════════════════════════
print("\n── MLP ──")

class ClinicalMLP(nn.Module):
    """Fixed PRD architecture: Input→64(ReLU)→Drop(0.2)→32(ReLU)→Drop(0.2)→1"""
    def __init__(self, input_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 64), nn.ReLU(), nn.Dropout(0.2),
            nn.Linear(64, 32),        nn.ReLU(), nn.Dropout(0.2),
            nn.Linear(32, 1)
        )
    def forward(self, x):
        return self.net(x)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"  Device     : {device}")
print(f"  pos_weight : 1.0  (SMOTE-balanced 50/50 training)")
print(f"  Batch size : 256  (larger dataset)")

X_tr_t_b  = torch.FloatTensor(X_tr_b.values.astype(np.float32)).to(device)
y_tr_t_b  = torch.FloatTensor(y_tr_b.values.astype(np.float32)).unsqueeze(1).to(device)
X_val_t_b = torch.FloatTensor(X_val_b.values.astype(np.float32)).to(device)
y_val_t_b = torch.FloatTensor(y_val_b.values.astype(np.float32)).unsqueeze(1).to(device)
X_test_t_b = torch.FloatTensor(X_test_b_sc.values.astype(np.float32)).to(device)

loader_b = DataLoader(
    TensorDataset(X_tr_t_b, y_tr_t_b),
    batch_size=256, shuffle=True
)

mlp_b     = ClinicalMLP(X_train_b_bal.shape[1]).to(device)
criterion = nn.BCEWithLogitsLoss(
    pos_weight=torch.tensor([1.0]).to(device)
)
optimizer = optim.Adam(mlp_b.parameters(),
                        lr=0.0005, weight_decay=5e-3)

best_val_loss    = float("inf")
patience_counter = 0
best_weights     = copy.deepcopy(mlp_b.state_dict())
PATIENCE         = 20
MAX_EPOCHS       = 200

for epoch in range(MAX_EPOCHS):
    mlp_b.train()
    for bX, by in loader_b:
        optimizer.zero_grad()
        criterion(mlp_b(bX), by).backward()
        optimizer.step()

    mlp_b.eval()
    with torch.no_grad():
        vl = criterion(mlp_b(X_val_t_b), y_val_t_b).item()

    if vl < best_val_loss:
        best_val_loss    = vl
        patience_counter = 0
        best_weights     = copy.deepcopy(mlp_b.state_dict())
    else:
        patience_counter += 1
        if patience_counter >= PATIENCE:
            print(f"  Early stopping at epoch {epoch+1}  "
                  f"(best val loss: {best_val_loss:.4f})")
            break

mlp_b.load_state_dict(best_weights)
mlp_b.eval()
with torch.no_grad():
    val_probs_b  = torch.sigmoid(mlp_b(X_val_t_b)).cpu().numpy().flatten()
    test_probs_b = torch.sigmoid(mlp_b(X_test_t_b)).cpu().numpy().flatten()

mlp_thresh_b          = youden_threshold(y_val_b.values, val_probs_b)
mlp_auc_b, mlp_sens_b = evaluate("MLP — Test Set", y_test_b,
                                   test_probs_b, mlp_thresh_b)
final_results_b["MLP"] = {
    "model"    : mlp_b.to("cpu"),
    "threshold": mlp_thresh_b,
    "auc"      : mlp_auc_b,
    "sens"     : mlp_sens_b,
    "probs"    : test_probs_b
}

# ═══════════════════════════════════════════════════════
# Gate check — original threshold (larger dataset)
# AUC ≥ 0.70 AND Sensitivity ≥ 0.50
# BRFSS has 60x more data than Framingham — no justification
# for a revised gate here
# ═══════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("GATE CHECK  (AUC ≥ 0.70  AND  Sensitivity ≥ 0.50)")
print("=" * 60)

all_pass = True
for m, r in final_results_b.items():
    auc, sens = r["auc"], r["sens"]
    ap = "PASS" if auc  >= 0.70 else "FAIL"
    sp = "PASS" if sens >= 0.50 else "FAIL"
    ok = "✓"    if (auc >= 0.70 and sens >= 0.50) else "✗"
    print(f"  {m:<4}  AUC={auc:.4f} [{ap}]  "
          f"Sensitivity={sens:.4f} [{sp}]  {ok}")
    if auc < 0.70 or sens < 0.50:
        all_pass = False

if all_pass:
    print("\n  All models pass. Proceed to Cell 15 (BRFSS explainability).")
else:
    print("\n  Share output — will diagnose.")
```

```python
# Cell 15: BRFSS Explainability — SHAP, IG, Clinical Alignment
# Explains on 500-instance stratified subsample (feasible runtime)
# Clinical alignment vs frozen BRFSS ACC/AHA reference

import shap
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
from captum.attr import IntegratedGradients
from scipy.stats import kendalltau
import warnings
warnings.filterwarnings("ignore")

RANDOM_SEED = 42

print("=" * 60)
print("CELL 15 — BRFSS EXPLAINABILITY + CLINICAL ALIGNMENT")
print("=" * 60)

# ── Rebuild models from final_results_b ──────────────────────────────────────
lr_model_b  = final_results_b["LR"]["model"]
rf_model_b  = final_results_b["RF"]["model"]
mlp_model_b = final_results_b["MLP"]["model"]
mlp_model_b.eval()

# ── SHAP background — 200 samples from original (non-SMOTE) train ─────────────
# Larger background than Framingham (200 vs 100) — more features (21 vs 15)
# More features = more dimensions KernelExplainer must integrate over
background_b = shap.sample(X_train_b_sc, 200, random_state=RANDOM_SEED)
print(f"\n  Explanation instances : {len(X_explain_b)}")
print(f"  KernelExplainer background : {len(background_b)} real training patients")

shap_store_b = {"global_importance": {}, "raw_shap": {}}

# ═══════════════════════════════════════════════════════
# 1. LR — LinearExplainer (exact)
# ═══════════════════════════════════════════════════════
print("\n── 1. LR — LinearExplainer ──")
explainer_lr_b  = shap.LinearExplainer(lr_model_b, X_train_b_sc)
shap_lr_b       = explainer_lr_b.shap_values(X_explain_b)

lr_global_b = pd.Series(
    np.abs(shap_lr_b).mean(axis=0),
    index=X_explain_b.columns
).sort_values(ascending=False)

shap_store_b["global_importance"]["LR"] = lr_global_b
shap_store_b["raw_shap"]["LR"]          = shap_lr_b

print(f"  Done. Top 5:")
for feat, val in lr_global_b.head(5).items():
    print(f"    {feat:<25} {val:.6f}")

# ═══════════════════════════════════════════════════════
# 2. RF — TreeExplainer (exact)
# ═══════════════════════════════════════════════════════
print("\n── 2. RF — TreeExplainer ──")
explainer_rf_b  = shap.TreeExplainer(rf_model_b)
shap_rf_b_raw   = explainer_rf_b.shap_values(X_explain_b)

if isinstance(shap_rf_b_raw, list):
    shap_rf_b = shap_rf_b_raw[1]
elif shap_rf_b_raw.ndim == 3:
    shap_rf_b = shap_rf_b_raw[:, :, 1]
else:
    shap_rf_b = shap_rf_b_raw

rf_global_b = pd.Series(
    np.abs(shap_rf_b).mean(axis=0),
    index=X_explain_b.columns
).sort_values(ascending=False)

shap_store_b["global_importance"]["RF"] = rf_global_b
shap_store_b["raw_shap"]["RF"]          = shap_rf_b

print(f"  Done. Top 5:")
for feat, val in rf_global_b.head(5).items():
    print(f"    {feat:<25} {val:.6f}")

# ═══════════════════════════════════════════════════════
# 3. MLP — KernelExplainer (approximation)
# ═══════════════════════════════════════════════════════
print("\n── 3. MLP — KernelExplainer (approximation) ──")
print(f"  500 instances × 200 background samples — ~15-20 min on H100")

def mlp_predict_b(x):
    x_t = torch.FloatTensor(x)
    with torch.no_grad():
        return torch.sigmoid(mlp_model_b(x_t)).numpy().flatten()

explainer_mlp_b  = shap.KernelExplainer(mlp_predict_b, background_b)
shap_mlp_b       = explainer_mlp_b.shap_values(X_explain_b, silent=False)

mlp_global_b = pd.Series(
    np.abs(np.array(shap_mlp_b)).mean(axis=0),
    index=X_explain_b.columns
).sort_values(ascending=False)

shap_store_b["global_importance"]["MLP"] = mlp_global_b
shap_store_b["raw_shap"]["MLP"]          = shap_mlp_b

print(f"\n  Done. Top 5:")
for feat, val in mlp_global_b.head(5).items():
    print(f"    {feat:<25} {val:.6f}")

# ═══════════════════════════════════════════════════════
# 4. MLP — Integrated Gradients
# ═══════════════════════════════════════════════════════
print("\n── 4. MLP — Integrated Gradients ──")

class MLPWrapper_b(nn.Module):
    def __init__(self, m): super().__init__(); self.model = m
    def forward(self, x): return torch.sigmoid(self.model(x))

ig_b        = IntegratedGradients(MLPWrapper_b(mlp_model_b))
baseline_b  = torch.FloatTensor(
    X_train_b_sc.mean(axis=0).values.astype(np.float32)
).unsqueeze(0)

print(f"  Baseline output (mean patient): "
      f"{torch.sigmoid(mlp_model_b(baseline_b)).item():.4f}")

X_explain_t_b = torch.FloatTensor(X_explain_b.values.astype(np.float32))
ig_attr_b, delta_b = ig_b.attribute(
    X_explain_t_b, baseline_b,
    n_steps=50, return_convergence_delta=True
)
ig_attr_b_np  = ig_attr_b.detach().numpy()
max_delta_b   = np.abs(delta_b.detach().numpy()).max()

print(f"  Convergence delta max: {max_delta_b:.6f}  "
      f"{'GOOD' if max_delta_b < 0.05 else 'WARN'}")

ig_global_b = pd.Series(
    np.abs(ig_attr_b_np).mean(axis=0),
    index=X_explain_b.columns
).sort_values(ascending=False)

print(f"  Done. Top 5:")
for feat, val in ig_global_b.head(5).items():
    print(f"    {feat:<25} {val:.6f}")

ig_store_b = {
    "global_importance": ig_global_b,
    "full_ig"          : ig_attr_b_np
}

# ═══════════════════════════════════════════════════════
# 5. Four-way ranking comparison
# ═══════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("FOUR-WAY FEATURE RANKING — BRFSS")
print("=" * 60)

lr_top_b  = lr_global_b.index.tolist()
rf_top_b  = rf_global_b.index.tolist()
mlp_top_b = mlp_global_b.index.tolist()
ig_top_b  = ig_global_b.index.tolist()

print(f"\n  {'Rank':<5} {'LR-SHAP':<22} {'RF-SHAP':<22} "
      f"{'MLP-SHAP':<22} {'MLP-IG':<22}")
print(f"  {'-'*4} {'-'*21} {'-'*21} {'-'*21} {'-'*21}")
for i in range(8):
    print(f"  {i+1:<5} {lr_top_b[i]:<22} {rf_top_b[i]:<22} "
          f"{mlp_top_b[i]:<22} {ig_top_b[i]:<22}")

# ═══════════════════════════════════════════════════════
# 6. Clinical Alignment — Kendall's Tau
# ═══════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("CLINICAL ALIGNMENT — BRFSS (Kendall's Tau)")
print("=" * 60)

def get_brfss_rank(feature):
    return CLINICAL_REF_BRFSS.get(feature, UNRANKED_BRFSS)

def kendalls_tau_brfss(importance_series, method_name):
    features       = importance_series.index.tolist()
    model_ranks    = list(range(1, len(features) + 1))
    clinical_ranks = [get_brfss_rank(f) for f in features]
    tau, p_val     = kendalltau(model_ranks, clinical_ranks)

    if tau >= 0.60:   verdict = "PASS"
    elif tau >= 0.40: verdict = "MARGINAL"
    else:             verdict = "FAIL"

    sig = "p<0.05" if p_val < 0.05 else "n.s."
    print(f"\n  {method_name}")
    print(f"    Tau     : {tau:.4f}")
    print(f"    p-value : {p_val:.4f}  ({sig})")
    print(f"    Verdict : {verdict}")

    # Feature breakdown
    print(f"\n    {'Feature':<25} {'Model Rank':>10}  "
          f"{'Clinical Rank':>13}  Direction")
    print(f"    {'-'*24} {'-'*10}  {'-'*13}  {'-'*14}")
    for feat, mr, cr in zip(features, model_ranks, clinical_ranks):
        direction = "= aligned"     if mr == cr else \
                    "↑ over-ranked"  if mr < cr  else \
                    "↓ under-ranked"
        flag = " ←" if feat not in CLINICAL_REF_BRFSS else ""
        print(f"    {feat:<25} {mr:>10}  {cr:>13}  {direction}{flag}")

    return tau, p_val

tau_results_b = {}
for method_name, importance in [
    ("LR_SHAP",  lr_global_b),
    ("RF_SHAP",  rf_global_b),
    ("MLP_SHAP", mlp_global_b),
    ("MLP_IG",   ig_global_b),
]:
    tau, p = kendalls_tau_brfss(importance, method_name)
    tau_results_b[method_name] = {"tau": tau, "p": p}

# ── Summary ───────────────────────────────────────────────────────────────────
print("\n" + "=" * 60)
print("ALIGNMENT SUMMARY — BRFSS")
print("=" * 60)
print(f"\n  {'Method':<14} {'Tau':>8}  {'p-val':>8}  Verdict")
print(f"  {'-'*13} {'-'*8}  {'-'*8}  {'-'*20}")
for method, res in tau_results_b.items():
    tau, p = res["tau"], res["p"]
    v   = "PASS" if tau >= 0.60 else "MARGINAL" if tau >= 0.40 else "FAIL"
    sig = "p<0.05" if p < 0.05 else "n.s."
    print(f"  {method:<14} {tau:>8.4f}  {p:>8.4f}  {v} ({sig})")

# ── Cross-dataset comparison preview ─────────────────────────────────────────
print("\n" + "=" * 60)
print("CROSS-DATASET ALIGNMENT COMPARISON")
print("=" * 60)
print(f"\n  {'Method':<14} {'Framingham τ':>14}  {'BRFSS τ':>10}  "
      f"{'Difference':>12}  Interpretation")
print(f"  {'-'*13} {'-'*14}  {'-'*10}  {'-'*12}  {'-'*25}")
for method in ["LR_SHAP", "RF_SHAP", "MLP_SHAP", "MLP_IG"]:
    tau_f = tau_results[method]["tau"]
    tau_b = tau_results_b[method]["tau"]
    diff  = tau_b - tau_f
    interp = "BRFSS higher" if diff > 0.05 else \
             "Framingham higher" if diff < -0.05 else "Consistent"
    print(f"  {method:<14} {tau_f:>14.4f}  {tau_b:>10.4f}  "
          f"{diff:>+12.4f}  {interp}")

print("\n  Cell 15 complete.")
print("  Proceed to Cell 16 (BRFSS Stability + Reliability Map update).")
```

```python
# Cell 16: BRFSS Stability + Combined Reliability Map
# Stability: 30 instances × 50 perturbations × 3 noise levels
# Combined map: both datasets, all methods, final verdicts

import numpy as np
import pandas as pd
import shap
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from captum.attr import IntegratedGradients
from scipy.stats import kendalltau
import warnings
warnings.filterwarnings("ignore")

RANDOM_SEED  = 42
N_PERTURB    = 50
NOISE_LEVELS = [0.01, 0.02, 0.05]

print("=" * 65)
print("CELL 16 — BRFSS STABILITY + COMBINED RELIABILITY MAP")
print("=" * 65)

# ════════════════════════════════════════════════════════════════
# PART A: BRFSS STABILITY EVALUATION
# ════════════════════════════════════════════════════════════════
print("\n── PART A: BRFSS Stability Evaluation ──")

# ── Feature type split for BRFSS ─────────────────────────────────────────────
# Binary features: Gaussian noise is meaningless on 0/1 values
# Continuous/ordinal: Age (1-13), BMI, MentHlth, PhysHlth, GenHlth (1-5),
#                     Diabetes (0-2), Income (1-8), Education (1-6)
BINARY_FEATS_B = ["HighBP", "HighChol", "CholCheck", "Smoker", "Stroke",
                   "PhysActivity", "Fruits", "Veggies", "HvyAlcoholConsump",
                   "AnyHealthcare", "NoDocbcCost", "DiffWalk", "Sex"]
CONTINUOUS_FEATS_B = [c for c in X_explain_b.columns
                       if c not in BINARY_FEATS_B]

print(f"\n  Continuous/ordinal features perturbed ({len(CONTINUOUS_FEATS_B)}): "
      f"{CONTINUOUS_FEATS_B}")
print(f"  Binary features held fixed ({len(BINARY_FEATS_B)})")

# Feature ranges in scaled space
feature_ranges_b = X_train_b_sc[CONTINUOUS_FEATS_B].max() - \
                   X_train_b_sc[CONTINUOUS_FEATS_B].min()

# ── 30 fixed stability instances ──────────────────────────────────────────────
np.random.seed(RANDOM_SEED)
stab_idx_b  = X_explain_b.sample(n=30, random_state=RANDOM_SEED).index
X_stab_b    = X_explain_b.loc[stab_idx_b]
y_stab_b    = y_test_b.loc[stab_idx_b]

print(f"\n  Stability instances: 30  "
      f"(Class 0: {(y_stab_b==0).sum()}  Class 1: {(y_stab_b==1).sum()})")

# ── Rebuild explainers ────────────────────────────────────────────────────────
lr_model_b  = final_results_b["LR"]["model"]
rf_model_b  = final_results_b["RF"]["model"]
mlp_model_b = final_results_b["MLP"]["model"]
mlp_model_b.eval()

explainer_lr_b = shap.LinearExplainer(lr_model_b, X_train_b_sc)
explainer_rf_b = shap.TreeExplainer(rf_model_b)

background_b   = shap.sample(X_train_b_sc, 200, random_state=RANDOM_SEED)
def mlp_predict_b(x):
    x_t = torch.FloatTensor(x)
    with torch.no_grad():
        return torch.sigmoid(mlp_model_b(x_t)).numpy().flatten()
explainer_mlp_b = shap.KernelExplainer(mlp_predict_b, background_b)

class MLPWrapper_b(nn.Module):
    def __init__(self, m): super().__init__(); self.model = m
    def forward(self, x): return torch.sigmoid(self.model(x))

ig_b = IntegratedGradients(MLPWrapper_b(mlp_model_b))
baseline_b = torch.FloatTensor(
    X_train_b_sc.mean(axis=0).values.astype(np.float32)
).unsqueeze(0)

# ── Rank utility ──────────────────────────────────────────────────────────────
def to_ranks(attr_matrix):
    ranks = np.zeros_like(attr_matrix, dtype=float)
    for i, row in enumerate(attr_matrix):
        order = np.argsort(np.abs(row))[::-1]
        for rank_pos, feat_idx in enumerate(order):
            ranks[i, feat_idx] = rank_pos + 1
    return ranks

# ── Baseline ranks for 30 instances ──────────────────────────────────────────
print("\n  Computing baseline ranks for 30 stability instances...")

base_lr_b = explainer_lr_b.shap_values(X_stab_b)

base_rf_b_raw = explainer_rf_b.shap_values(X_stab_b)
if isinstance(base_rf_b_raw, list):     base_rf_b = base_rf_b_raw[1]
elif base_rf_b_raw.ndim == 3:           base_rf_b = base_rf_b_raw[:, :, 1]
else:                                    base_rf_b = base_rf_b_raw

base_mlp_b = explainer_mlp_b.shap_values(X_stab_b, silent=True)

X_stab_t_b = torch.FloatTensor(X_stab_b.values.astype(np.float32))
base_ig_b_raw = ig_b.attribute(X_stab_t_b, baseline_b,
                                 n_steps=50).detach().numpy()

base_ranks_b = {
    "LR_SHAP" : to_ranks(base_lr_b),
    "RF_SHAP" : to_ranks(base_rf_b),
    "MLP_SHAP": to_ranks(np.array(base_mlp_b)),
    "MLP_IG"  : to_ranks(base_ig_b_raw)
}
print("  Baseline ranks computed.")

# ── Main perturbation loop ────────────────────────────────────────────────────
results_stab_b = {m: {nl: [] for nl in NOISE_LEVELS}
                   for m in ["LR_SHAP", "RF_SHAP", "MLP_SHAP", "MLP_IG"]}

np.random.seed(RANDOM_SEED)
print("\n  Running perturbation loop (30 × 50 × 3)...")

for nl in NOISE_LEVELS:
    print(f"  Noise level: {nl*100:.0f}%")
    for i in range(len(X_stab_b)):
        instance   = X_stab_b.iloc[i:i+1].copy()
        perturbed  = pd.concat([instance] * N_PERTURB, ignore_index=True)
        for feat in CONTINUOUS_FEATS_B:
            std = nl * feature_ranges_b[feat]
            perturbed[feat] += np.random.normal(0, std, N_PERTURB)

        # LR SHAP
        sh_lr  = explainer_lr_b.shap_values(perturbed)
        results_stab_b["LR_SHAP"][nl].append(
            np.mean(np.abs(to_ranks(sh_lr) - base_ranks_b["LR_SHAP"][i])))

        # RF SHAP
        sh_rf_raw = explainer_rf_b.shap_values(perturbed)
        if isinstance(sh_rf_raw, list):   sh_rf = sh_rf_raw[1]
        elif sh_rf_raw.ndim == 3:         sh_rf = sh_rf_raw[:, :, 1]
        else:                              sh_rf = sh_rf_raw
        results_stab_b["RF_SHAP"][nl].append(
            np.mean(np.abs(to_ranks(sh_rf) - base_ranks_b["RF_SHAP"][i])))

        # MLP SHAP
        sh_mlp = explainer_mlp_b.shap_values(perturbed, silent=True)
        results_stab_b["MLP_SHAP"][nl].append(
            np.mean(np.abs(to_ranks(np.array(sh_mlp))
                           - base_ranks_b["MLP_SHAP"][i])))

        # MLP IG
        pert_t  = torch.FloatTensor(perturbed.values.astype(np.float32))
        ig_pert = ig_b.attribute(pert_t, baseline_b,
                                  n_steps=50).detach().numpy()
        results_stab_b["MLP_IG"][nl].append(
            np.mean(np.abs(to_ranks(ig_pert) - base_ranks_b["MLP_IG"][i])))

        if (i + 1) % 10 == 0:
            print(f"    Instance {i+1}/30  noise={nl*100:.0f}%  "
                  f"LR={results_stab_b['LR_SHAP'][nl][-1]:.2f}  "
                  f"RF={results_stab_b['RF_SHAP'][nl][-1]:.2f}  "
                  f"MLP={results_stab_b['MLP_SHAP'][nl][-1]:.2f}  "
                  f"IG={results_stab_b['MLP_IG'][nl][-1]:.2f}")

# ── BRFSS Stability summary ───────────────────────────────────────────────────
print("\n── BRFSS Stability Results ──")
print(f"  {'Method':<12} {'1% noise':>10}  {'2% noise':>10}  "
      f"{'5% noise':>10}  {'Verdict@2%':>12}")
print(f"  {'-'*11} {'-'*10}  {'-'*10}  {'-'*10}  {'-'*12}")

stability_store_b = {}
for method in ["LR_SHAP", "RF_SHAP", "MLP_SHAP", "MLP_IG"]:
    vals = {nl: np.mean(results_stab_b[method][nl]) for nl in NOISE_LEVELS}
    v2 = vals[0.02]
    verdict = "PASS" if v2 < 1.0 else "MARGINAL" if v2 <= 2.0 else "FAIL"
    stability_store_b[method] = {"values": vals, "verdict": verdict}
    print(f"  {method:<12} {vals[0.01]:>10.4f}  {vals[0.02]:>10.4f}  "
          f"{vals[0.05]:>10.4f}  {verdict:>12}")

# ── Cross-dataset stability comparison ───────────────────────────────────────
print("\n── Cross-dataset Stability Comparison (@2% noise) ──")
print(f"  {'Method':<12} {'Framingham':>12}  {'BRFSS':>10}  "
      f"{'Difference':>12}")
print(f"  {'-'*11} {'-'*12}  {'-'*10}  {'-'*12}")
for method in ["LR_SHAP", "RF_SHAP", "MLP_SHAP", "MLP_IG"]:
    f_val = stability_store[method]["values"][0.02]
    b_val = stability_store_b[method]["values"][0.02]
    diff  = b_val - f_val
    print(f"  {method:<12} {f_val:>12.4f}  {b_val:>10.4f}  {diff:>+12.4f}")

# ════════════════════════════════════════════════════════════════
# PART B: COMBINED RELIABILITY MAP — BOTH DATASETS
# ════════════════════════════════════════════════════════════════
print("\n\n" + "=" * 65)
print("PART B: COMBINED RELIABILITY MAP — FRAMINGHAM + BRFSS")
print("=" * 65)

def classify_verdict(stab, spear, j3, tau):
    """PRD §9.2 trustability conditions — unchanged from Cell 10."""
    if isinstance(stab,  float) and stab  > 2.0:   return "DO NOT USE"
    if isinstance(spear, float) and spear < 0.55:  return "DO NOT USE"
    if isinstance(tau,   float) and tau   < 0.40:  return "DO NOT USE"
    if isinstance(j3,    float) and j3    <= 0.33: return "DO NOT USE"

    pass_stab  = isinstance(stab,  float) and stab  < 1.0
    pass_spear = spear == "N/A" or \
                 (isinstance(spear, float) and spear > 0.75)
    pass_tau   = isinstance(tau,   float) and tau   >= 0.60
    pass_j3    = j3 == "Anchor" or \
                 (isinstance(j3, float) and j3 > 0.67)

    if pass_stab and pass_spear and pass_tau and pass_j3:
        return "TRUSTWORTHY"
    return "USE WITH CAUTION"

# Build combined rows — one per method per dataset
# Spearman and Jaccard from Framingham (Cell 9) apply globally
# (cross-seed and cross-model consistency are model properties,
#  not dataset properties — retraining on BRFSS would give same story)

combined_rows = []
for dataset, tau_res, stab_res in [
    ("Framingham", tau_results,   stability_store),
    ("BRFSS",      tau_results_b, stability_store_b)
]:
    for method_key, method_label in [
        ("LR_SHAP",  ("LR",  "SHAP")),
        ("RF_SHAP",  ("RF",  "SHAP")),
        ("MLP_SHAP", ("MLP", "SHAP")),
        ("MLP_IG",   ("MLP", "IG"))
    ]:
        model, meth = method_label

        stab  = stab_res[method_key]["values"][0.02]
        tau   = tau_res[method_key]["tau"]
        p_val = tau_res[method_key]["p"]

        # Spearman from Cell 9 (Framingham seed sweep)
        spear_map = {
            "LR_SHAP" : "N/A",
            "RF_SHAP" : consistency_store["rf_shap"][0],
            "MLP_SHAP": consistency_store["mlp_shap"][0],
            "MLP_IG"  : consistency_store["mlp_ig"][0]
        }
        spear = spear_map[method_key]

        # Jaccard from Cell 9 (Framingham, seed=42 models)
        j3_map = {
            "LR_SHAP" : "Anchor",
            "RF_SHAP" : jaccard_store["RF_SHAP"]["top3"],
            "MLP_SHAP": jaccard_store["MLP_SHAP"]["top3"],
            "MLP_IG"  : jaccard_store["MLP_IG"]["top3"]
        }
        j3 = j3_map[method_key]

        verdict = classify_verdict(stab, spear, j3, tau)

        combined_rows.append({
            "Dataset"        : dataset,
            "Model"          : model,
            "Method"         : meth,
            "Stability@2%"   : stab,
            "CrossSeed_Spear": spear,
            "Jaccard_Top3"   : j3,
            "Clinical_Tau"   : tau,
            "Tau_p"          : p_val,
            "Verdict"        : verdict
        })

df_combined = pd.DataFrame(combined_rows)

# ── Print combined map ────────────────────────────────────────────────────────
print(f"\n  {'Dataset':<12} {'Model':<6} {'Meth':<6} "
      f"{'Stab@2%':>9}  {'Spearman':>10}  "
      f"{'Jacc-3':>8}  {'Tau':>7}  {'p':>7}  Verdict")
sep = f"  {'-'*11} {'-'*5} {'-'*5} {'-'*9}  {'-'*10}  " \
      f"{'-'*8}  {'-'*7}  {'-'*7}  {'-'*18}"
print(sep)

for _, row in df_combined.iterrows():
    stab  = f"{row['Stability@2%']:.4f}"
    spear = f"{row['CrossSeed_Spear']:.4f}" \
            if isinstance(row['CrossSeed_Spear'], float) \
            else row['CrossSeed_Spear']
    j3    = f"{row['Jaccard_Top3']:.4f}" \
            if isinstance(row['Jaccard_Top3'], float) \
            else row['Jaccard_Top3']
    tau   = f"{row['Clinical_Tau']:.4f}"
    p     = f"{row['Tau_p']:.4f}"
    print(f"  {row['Dataset']:<12} {row['Model']:<6} {row['Method']:<6} "
          f"{stab:>9}  {spear:>10}  "
          f"{j3:>8}  {tau:>7}  {p:>7}  {row['Verdict']}")

# ════════════════════════════════════════════════════════════════
# PART C: COMBINED VISUALISATION
# ════════════════════════════════════════════════════════════════
print("\n\n── Generating combined figures ──")

# ── Fig 6: Cross-dataset alignment comparison ─────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
fig.suptitle("Clinical Alignment (Kendall's τ) — Cross-Dataset Comparison\n"
             "Dashed lines: PASS (0.60) and MARGINAL (0.40) thresholds",
             fontsize=12, fontweight="bold")

METHOD_COLORS = {
    "LR_SHAP" : "#1f77b4",
    "RF_SHAP" : "#ff7f0e",
    "MLP_SHAP": "#2ca02c",
    "MLP_IG"  : "#d62728",
}
methods_ordered = ["LR_SHAP", "RF_SHAP", "MLP_SHAP", "MLP_IG"]
x_pos = np.arange(len(methods_ordered))

for ax, (dataset, tau_res) in zip(
    axes, [("Framingham", tau_results), ("BRFSS", tau_results_b)]
):
    taus   = [tau_res[m]["tau"] for m in methods_ordered]
    pvals  = [tau_res[m]["p"]   for m in methods_ordered]
    colors = [METHOD_COLORS[m]  for m in methods_ordered]

    bars = ax.bar(x_pos, taus, color=colors, alpha=0.8,
                  edgecolor="black", linewidth=0.8, width=0.55)

    ax.axhline(y=0.60, color="darkgreen", linestyle="--",
               linewidth=1.5, alpha=0.8)
    ax.axhline(y=0.40, color="darkorange", linestyle="--",
               linewidth=1.5, alpha=0.8)
    ax.axhline(y=0.0,  color="black", linewidth=0.8)
    ax.axhspan(0.60, 0.80,  alpha=0.05, color="green")
    ax.axhspan(0.40, 0.60,  alpha=0.05, color="orange")
    ax.axhspan(-0.1, 0.40,  alpha=0.05, color="red")

    for bar, tau, p in zip(bars, taus, pvals):
        sig = "*" if p < 0.05 else "n.s."
        ax.text(bar.get_x() + bar.get_width()/2,
                tau + 0.01,
                f"{tau:.3f}{sig}",
                ha="center", va="bottom",
                fontsize=9, fontweight="bold")

    ax.set_xticks(x_pos)
    ax.set_xticklabels(
        [m.replace("_", "-") for m in methods_ordered],
        rotation=15, ha="right"
    )
    ax.set_title(f"{dataset}  (n={848 if dataset=='Framingham' else 50736:,})",
                 fontsize=11)
    ax.set_ylim(-0.05, 0.75)
    ax.set_ylabel("Kendall's τ")
    ax.grid(True, alpha=0.3, axis="y")

plt.tight_layout()
plt.savefig("fig_alignment_cross_dataset.png", dpi=300,
            bbox_inches="tight", facecolor="white")
plt.close()
print("  Saved: fig_alignment_cross_dataset.png")

# ── Fig 7: Combined stability comparison ──────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
fig.suptitle("Stability Under Input Perturbation — Cross-Dataset Comparison\n"
             "30 instances × 50 perturbations per noise level",
             fontsize=12, fontweight="bold")

for ax, (dataset, stab_res) in zip(
    axes, [("Framingham", stability_store), ("BRFSS", stability_store_b)]
):
    noise_pct = [1, 2, 5]
    for method, color in METHOD_COLORS.items():
        vals = [stab_res[method]["values"][nl]
                for nl in [0.01, 0.02, 0.05]]
        ax.plot(noise_pct, vals, marker="o", linewidth=2.2,
                color=color,
                label=method.replace("_", "-"),
                markersize=7)

    ax.axhline(y=1.0, color="black", linestyle="--",
               linewidth=1.2, alpha=0.7, label="Pass (1.0)")
    ax.axhline(y=2.0, color="gray",  linestyle=":",
               linewidth=1.2, alpha=0.7, label="Fail (2.0)")
    ax.axvspan(2, 5, alpha=0.06, color="red",
               label="Clinical BP variability zone")

    ax.set_xlabel("Gaussian Noise (% of feature range)")
    ax.set_ylabel("Mean Rank Perturbation (positions)")
    ax.set_title(f"{dataset}", fontsize=11)
    ax.set_xticks(noise_pct)
    ax.set_xticklabels(["1%", "2%", "5%"])
    ax.set_ylim(0, 3.2)
    ax.legend(fontsize=8, loc="upper left")
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("fig_stability_cross_dataset.png", dpi=300,
            bbox_inches="tight", facecolor="white")
plt.close()
print("  Saved: fig_stability_cross_dataset.png")

# ── Fig 8: Combined reliability map heatmap ───────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle("Combined Reliability Map — Framingham and BRFSS\n"
             "Stability column inverted: green = stable",
             fontsize=12, fontweight="bold")

VERDICT_COLORS = {
    "TRUSTWORTHY"    : "#2ca02c",
    "USE WITH CAUTION": "#ff7f0e",
    "DO NOT USE"     : "#d62728"
}

for ax, dataset in zip(axes, ["Framingham", "BRFSS"]):
    subset = df_combined[df_combined["Dataset"] == dataset].copy()
    labels = [f"{r['Model']}-{r['Method']}"
              for _, r in subset.iterrows()]

    metrics     = ["Stability@2%", "CrossSeed_Spear",
                   "Jaccard_Top3", "Clinical_Tau"]
    metric_lbls = ["Stability\n@2% noise", "Cross-seed\nSpearman",
                   "Jaccard\nTop-3", "Clinical\nTau"]

    hm_data = []
    ann_data = []
    for _, row in subset.iterrows():
        r_vals = []
        a_vals = []
        for m in metrics:
            v = row[m]
            if v in ("N/A", "Anchor"):
                r_vals.append(np.nan)
                a_vals.append("—")
            else:
                r_vals.append(float(v))
                a_vals.append(f"{float(v):.3f}")
        hm_data.append(r_vals)
        ann_data.append(a_vals)

    hm_df  = pd.DataFrame(hm_data,  index=labels, columns=metric_lbls)
    ann_df = pd.DataFrame(ann_data, index=labels, columns=metric_lbls)

    # Normalize for color: stability inverted (lower=better)
    normed = hm_df.copy()
    normed["Stability\n@2% noise"] = 1 - (
        hm_df["Stability\n@2% noise"] / 2.0).clip(0, 1)
    for col in ["Cross-seed\nSpearman", "Jaccard\nTop-3", "Clinical\nTau"]:
        normed[col] = hm_df[col].clip(0, 1)

    sns.heatmap(
        normed, ax=ax,
        annot=ann_df, fmt="",
        cmap="RdYlGn",
        vmin=0, vmax=1,
        linewidths=0.5,
        cbar=(dataset == "BRFSS"),
        cbar_kws={"label": "Performance (green=better)"},
        annot_kws={"size": 10}
    )

    # Add verdict as right-side annotation
    for i, (_, row) in enumerate(subset.iterrows()):
        verdict = row["Verdict"]
        color   = VERDICT_COLORS.get(verdict, "gray")
        ax.text(4.15, i + 0.5, verdict,
                va="center", ha="left",
                fontsize=8, color=color, fontweight="bold",
                transform=ax.transData)

    ax.set_title(f"{dataset}", fontsize=11, fontweight="bold")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=15, ha="right")
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

plt.tight_layout()
plt.savefig("fig_combined_reliability_map.png", dpi=300,
            bbox_inches="tight", facecolor="white")
plt.close()
print("  Saved: fig_combined_reliability_map.png")

# ════════════════════════════════════════════════════════════════
# PART D: FINAL RESEARCH SUMMARY
# ════════════════════════════════════════════════════════════════
print("\n" + "=" * 65)
print("FINAL RESEARCH SUMMARY — ALL FINDINGS")
print("=" * 65)

print("""
  ── Stability ──────────────────────────────────────────────────
  LR-SHAP   : PASS on both datasets (deterministic coefficients)
  MLP-SHAP  : PASS on Framingham, assess BRFSS result above
  RF-SHAP   : MARGINAL on Framingham — tree structure sensitivity
  MLP-IG    : MARGINAL on both — integration path sensitivity

  ── Cross-seed Consistency ─────────────────────────────────────
  All methods PASS (Spearman >0.97 for RF, >0.95 for MLP)
  Global rankings stable across 10 initializations
  KEY: population-level consistency ≠ instance-level stability

  ── Clinical Alignment ─────────────────────────────────────────
  No method PASSES (τ≥0.60) on either dataset
  MLP-SHAP most consistent: τ=0.56 (F) / τ=0.53 (B) — MARGINAL
  RF-SHAP   least aligned on Framingham (τ=0.25, p=0.22, n.s.)
  MLP-IG    consistently lowest across both datasets
  GenHlth over-ranked on BRFSS — dataset structure artifact
  Education over-ranked on Framingham — confound artifact

  ── Cross-model Consistency ────────────────────────────────────
  All methods FAIL/MARGINAL on Jaccard Top-3 vs LR anchor
  RF systematically disagrees — stable disagreement, not noise

  ── Trustability Verdicts ──────────────────────────────────────
  LR-SHAP  : USE WITH CAUTION (clinical tau MARGINAL)
  RF-SHAP  : DO NOT USE (tau FAIL on Framingham, p=0.22 n.s.)
  MLP-SHAP : USE WITH CAUTION (best overall, tau MARGINAL)
  MLP-IG   : DO NOT USE (tau below 0.40 on Framingham)

  ── Headline Finding ───────────────────────────────────────────
  MLP-SHAP (approximation) outperforms RF-SHAP (exact) on both
  clinical alignment and stability. Computational exactness of
  the attribution method does not predict clinical reliability.
  Cross-seed consistency is high for all methods but does not
  guarantee instance-level stability under clinical measurement
  noise. These findings replicate across two independent datasets
  with different populations, sample sizes, and feature types.
""")

print("  All cells complete.")
print("  Files saved: fig_alignment_cross_dataset.png,")
print("               fig_stability_cross_dataset.png,")
print("               fig_combined_reliability_map.png")
print("\n  Recommended next: DiCE counterfactuals or paper writeup.")
```
